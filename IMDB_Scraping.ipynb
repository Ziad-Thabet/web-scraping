```python
import requests
from bs4 import BeautifulSoup
import csv
import time
```

```python
url = "https://m.imdb.com/chart/top/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
response = requests.get(url, headers=headers)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
```

```python
movie_urls = []
for movie_link in soup.find_all("a", {"class": "ipc-lockup-overlay"}):
    movie_url = "https://m.imdb.com" + movie_link["href"]
    movie_urls.append(movie_url)
movie_urls
```

```python
movie_urls = movie_urls[:-1]
movie_urls
```

```python
with open('movies_data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Director', 'Stars', 'Rating', 'Genre', 'Release Year','Image'])

    for url in movie_urls:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        title = soup.find("span", {'class': 'hero__primary-text'}).text
        director = soup.find("a", {"class": "ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"}).text
        stars = [star.text for star in soup.find_all("ul", {"class": "ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt"})[1:]]
        rating = soup.find("div", {"data-testid": "hero-rating-bar__aggregate-rating__score"}).text
        genre = [genre.text for genre in soup.find_all("a", {"class": "ipc-chip ipc-chip--on-baseAlt"})]
        release_year_text = soup.find("ul", {"class": "ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt"}).text
        release_year = release_year_text[:4]
        image_url = soup.find('img', {'class': 'ipc-image'})['src']

        writer.writerow([title, director, ', '.join(stars), rating, ', '.join(genre), release_year,image_url])
        print(f"{title} saved successfully.")
print("All movies have been saved.")
```
